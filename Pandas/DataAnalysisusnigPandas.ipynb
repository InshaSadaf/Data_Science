```python
def runs_score(batsman_name):
    vk = delivery[delivery['batsman']==batsman_name]
    runs = vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(1)
    return runs
runs_score('SK Raina')
```

```python
#isin function
# find the most destructive death over batsman in the hstory of IPL
#strike rate = (no of runs/no of balls)*100
death_over = delivery[delivery['over']>15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsman >200
all_batsman[x].shape
```

```python
all_batsman[x].index
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
#runs score by all 43 batsman
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```
