```python
#label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#colors and lines 
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 3)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linewidth = 2)
plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#to see the points
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 3,marker = 'D',markersize = 3)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linewidth = 2,marker ='o')
plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#legend -> location 
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 3,marker = 'D',markersize = 3,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linewidth = 2,marker ='o',label = 'Rohit')
plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')


# to see the labels on the graph we use
plt.legend(loc ='best')
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2017'],label ='2015')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom = df['2017'] + df['2016'],label = '2017')

plt.legend()
```

```python
gl = pd.read_csv('gayle-175.csv')
```

```python
gl
```

```python
plt.pie(gl['batsman_runs'],labels = gl['batsman'],autopct = '%0.1f%%')
plt.show()
```

```python
#colors
plt.pie(gl['batsman_runs'],labels = gl['batsman'],autopct = '%0.1f%%',colors = ['blue','pink','yellow','green','brown','purple'])
plt.show()
```

```python
#explode shadow
plt.pie(gl['batsman_runs'],labels = gl['batsman'],autopct = '%0.1f%%',colors = ['Skyblue','pink','yellow','green','brown','purple'],explode = [0.1,0,0,0,0.3,0])
plt.show()
```

```python
#shadow
#explode shadow
plt.pie(gl['batsman_runs'],labels = gl['batsman'],autopct = '%0.1f%%',colors = ['Skyblue','pink','yellow','green','brown','purple'],explode = [0.1,0,0,0,0.3,0],shadow = True)
plt.show()
```
